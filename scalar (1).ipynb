```python
# lakukan tahapan pada soal 1 dengan menggunakan tensorflow
import tensorflow as tf
```

```python
x_tf = tf.Variable(1089, dtype=tf.int16) # dtype is optional
x_tf
```

```python
x_tf.shape
```

```python
y_tf = tf.Variable(89, dtype=tf.int16)
```

```python
x_tf + y_tf
```

```python
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
tf_sum.numpy() # note that NumPy operations automatically convert tensors to NumPy arrays, and vice versa
```

```python
type(tf_sum.numpy())
```
